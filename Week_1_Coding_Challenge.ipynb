We can also calculate the area of a triangle using the Heron’s formula as follows.

Write a program that will output the area of a triangle given the side lengths. 

Here a few sample runs.

A: 10

A: 3
B: 4
C: 5
The area is 6.0.

A: 10.5

The area is 81.66911135796691.

```python
import math

A = float(input("A: "))
B = float(input("B: "))
C = float(input("C: "))
S = (A + B + C) / 2
area = math.sqrt(S * (S - A) * (S - B) * (S - C))
print("The area is "+str(area)+".")
```
